```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pre-processing step

Please make sure to fill in your matricule `mat`

```python
data = pd.read_csv("../data/ssh3201_intra_a22.csv", sep = ";", index_col = 0)
data["Q1/7"] = data["Q1/7"].str.replace(",", ".").astype(float)
data["Q2/6"] = data["Q2/6"].str.replace(",", ".").astype(float)
data["Q3/7"] = data["Q3/7"].str.replace(",", ".").astype(float)
data["Note/20"] = data["Note/20"].str.replace(",", ".").astype(float)

mat = 2232112
grade = data.loc[mat]
print(f"Grade details:\n{grade}")
```

## Visualization step

```python
data.describe()
```

### Notes Intra

```python
sns.histplot(data["Note/20"], kde = True)
plt.axvline(x = grade["Note/20"], c = "r", linestyle = "dashed")
plt.title("Note Intra")
plt.plot()
```

```python
sns.boxplot(data["Note/20"])
plt.axvline(x = grade["Note/20"], c = "r", linestyle = "dashed")
plt.title("Note Intra")
plt.plot()
```

```python
p = np.sum(data["Note/20"] >= grade["Note/20"]) / data.shape[0]
print(f"You are in the top {np.round(p, 3) * 100} % ")
```

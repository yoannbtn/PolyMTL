```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pre-processing step

Please make sure to fill in your matricule `mat`

```python
data = pd.read_csv("../data/inf8111_intra_a22.csv", sep = ";", index_col = 0)
data["Q1/2"] = data["Q1/2"].str.replace(",", ".").astype(float)
data["Q2/2"] = data["Q2/2"].str.replace(",", ".").astype(float)
data["Q3/2"] = data["Q3/2"].str.replace(",", ".").astype(float)
data["Q4/2"] = data["Q4/2"].str.replace(",", ".").astype(float)
data["Q5/2"] = data["Q5/2"].str.replace(",", ".").astype(float)
data["Global/10"] = data["Global/10"].str.replace(",", ".").astype(float)
data["Global arrondi/10"] = data["Global arrondi/10"].str.replace(",", ".").astype(float)

mat = 2232112
grade = data.loc[mat]
print(f"Grade details:\n{grade}")
```

## Visualization step

```python
data.describe()
```

### Notes globales arrondies

```python
sns.histplot(data["Global arrondi/10"], kde = True)
plt.axvline(x = grade["Global arrondi/10"], c = "r", linestyle = "dashed")
plt.title("Notes globales arrondies")
plt.plot()
```

```python
sns.boxplot(data["Global arrondi/10"])
plt.axvline(x = grade["Global arrondi/10"], c = "r", linestyle = "dashed")
plt.title("Notes globales arrondies")
plt.plot()
```

```python
p = np.sum(data["Global arrondi/10"] >= grade["Global arrondi/10"]) / data.shape[0]
print(f"You are in the top {np.round(p, 3) * 100} % ")
```

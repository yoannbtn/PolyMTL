```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pre-processing step

Please make sure to fill in your matricule `mat`

```python
data = pd.read_csv("../data/ssh3201_tps_a22.csv", sep = ";", index_col = 0)
data["TP1"] = data["TP1"].str.replace(",", ".").astype(float)
data["TP2"] = data["TP2"].str.replace(",", ".").astype(float)
data["TP3"] = data["TP3"].str.replace(",", ".").astype(float)
data["TP4"] = data["TP4"].str.replace(",", ".").astype(float)
data["TP5"] = data["TP5"].str.replace(",", ".").astype(float)
data["Moyenne"] = data.mean(axis = 1)

mat = 2232112
grade = data.loc[mat]
print(f"Grade details:\n{grade}")
```

## Visualization step

```python
data.describe()
```

### Notes TP1

```python
sns.histplot(data["TP1"], kde = True)
plt.axvline(x = grade["TP1"], c = "r", linestyle = "dashed")
plt.title("Note TP1")
plt.plot()
```

```python
sns.boxplot(data["TP1"])
plt.axvline(x = grade["TP1"], c = "r", linestyle = "dashed")
plt.title("Note TP1")
plt.plot()
```

```python
p = np.sum(data["TP1"] >= grade["TP1"]) / data.shape[0]
print(f"You are in the top {np.round(p, 3) * 100} % ")
```
